```python
spreadsheet = gc.open('Demark Indicators')
dfs = []
df = pd.DataFrame()

# Iterate through all sheets in the Google Sheets document
for worksheet in spreadsheet.worksheets():
    sheet_name = worksheet.title

    # Skip the sheet named "Sheet1"
    if sheet_name == 'Sheet1':
        continue

    # Read the data from the current sheet
    data = worksheet.get_all_values()

    # Convert the data to a DataFrame and do symbol level calculations
    df = pd.DataFrame(data[1:], columns=data[0])
    df.reset_index(drop=True, inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values(by='Date', inplace=True)
    df['Close_3_periods'] = df['Close'].shift(-3)
    df['Close_7_periods'] = df['Close'].shift(-7)
    df['Close_14_periods'] = df['Close'].shift(-14)
    df['Close_21_periods'] = df['Close'].shift(-21)

    print(sheet_name, len(df))

    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames in the list into one final DataFrame
final_df = pd.concat(dfs, ignore_index=True)
print(final_df.head())
```

## Analysis and Results

```python
# df loc to split up and do analysis
sell_df = df[df['Demark 13 Countdown Sell'] == 13].reset_index()
buy_df = df[df['Demark 13 Countdown Buy'] == 13].reset_index()

# return calculations - adding a buffer to account for transaction costs and whatnot
buy_df['Return_3_Periods'] = (buy_df['Close_3_periods'] - buy_df['Close']) / buy_df['Close']
buy_df['Return_7_Periods'] = (buy_df['Close_7_periods'] - buy_df['Close']) / buy_df['Close']
buy_df['Return_14_Periods'] = (buy_df['Close_14_periods'] - buy_df['Close']) / buy_df['Close']
buy_df['Return_21_Periods'] = (buy_df['Close_21_periods'] - buy_df['Close']) / buy_df['Close']
buy_df['day_of_week'] = buy_df['Date'].dt.dayofweek  # Monday=0, Sunday=6


sell_df['Return_3_Periods'] = (sell_df['Close'] - sell_df['Close_3_periods']) / sell_df['Close']
sell_df['Return_7_Periods'] = (sell_df['Close'] - sell_df['Close_7_periods']) / sell_df['Close']

sell_df['Return_14_Periods'] = (sell_df['Close'] - sell_df['Close_14_periods']) / sell_df['Close']
sell_df['Return_21_Periods'] = (sell_df['Close'] - sell_df['Close_21_periods']) / sell_df['Close']
sell_df['day_of_week'] = sell_df['Date'].dt.dayofweek  # Monday=0, Sunday=6

stats_by_day_sell = sell_df.groupby('day_of_week')['Return_3_Periods'].agg(['mean', 'count'])
stats_by_day_buy = buy_df.groupby('day_of_week')['Return_3_Periods'].agg(['mean', 'count'])


print("Rows In Sell_df: ", len(sell_df) )
print("Rows In Buy_df: ", len(buy_df) )
print("Sell Stats by end of the week")
print(stats_by_day_sell.head(5))

print("Buy Stats by end of the week")
print(stats_by_day_buy.head(5))
```

```python
# roi by time period
sns.set_style('darkgrid')

fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharey=True)
axs = axs.flatten()
# List of periods
periods = ['Return_3_Periods', 'Return_7_Periods', 'Return_14_Periods', 'Return_21_Periods']

colors = ['#05878a', '#67074e']
# Iterate over the periods and create a plot for each one
for i, period in enumerate(periods):
    # Separate the data into two groups
    positive = sell_df[sell_df[period] > 0.02][period]
    negative = sell_df[sell_df[period] < 0.02][period]

    # Plot the positive data in green
    sns.histplot(positive, kde=True, color=colors[0], ax=axs[i])

    # Plot the negative data in red
    sns.histplot(negative, kde=True, color=colors[1], ax=axs[i])

    # Set the title and labels
    axs[i].set_title(f'Density Plot of {period}')
    axs[i].set_xlabel('Return')
    axs[i].set_ylabel('Count Of Events')

# Adjust the layout
plt.tight_layout()

# Display the plots
plt.show()
print("Demark Sell Results")

# List of periods
periods = ['Return_3_Periods', 'Return_7_Periods', 'Return_14_Periods', 'Return_21_Periods']

# Initialize an empty DataFrame to store the results
results = pd.DataFrame(index=periods, columns=['% Above 2%', 'Mean(%)', 'Median(%)', 'St Dev', 'Sharpe', 'Sortino'])

# Iterate over the periods and calculate the percentage of values above 0 and the mean
for period in periods:
    results.loc[period, '% Above 2%'] = round((sell_df[period] > 0.02).mean() * 100,1)
    results.loc[period, 'Mean(%)'] = round(sell_df[period].mean()*100, 1)
    results.loc[period, 'Median(%)'] = round(sell_df[period].median()*100, 1)
    results.loc[period, 'St Dev'] = round(sell_df[period].std()*100, 1)
    results.loc[period, 'Sharpe'] = round((sell_df[period].mean() - 0.01 )/sell_df[period].std(), 1)
    results.loc[period, 'Sortino'] = round((sell_df[period].mean() - 0.01 )/(sell_df[period].min()), 2)


# Print the results
print(results)
```
